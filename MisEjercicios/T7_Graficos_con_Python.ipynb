```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(1, figsize=[15, 10])

n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint = True)
Y = np.sin(2 * X)
plt.plot(X, Y + 1, color = "blue", alpha = 1.00)
plt.plot(X, Y - 1, color = "blue", alpha = 1.00)

plt.fill_between(X, 1, Y+1, color = "blue", alpha = 0.2)
plt.fill_between(X, -1, Y-1, where = Y > 0, color = "blue", alpha = 0.2)
plt.fill_between(X, -1, Y-1, where = Y < 0, color = "red", alpha = 0.2)

plt.show()
```

```python
plt.figure(2, figsize=[15, 10])

n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
plt.scatter(X, Y, edgecolors = "black", linewidth = 1, c=np.arctan2(X, Y), cmap='rainbow', s = 100, alpha = 0.6)

plt.show()
```

```python
plt.figure(3, figsize=[15, 10])

n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
plt.bar(X, +Y1, facecolor = "#9999ff", edgecolor = "white")
plt.bar(X, -Y2, facecolor = "#ff9999", edgecolor = "white")

for x, y in zip(X, Y1):
    plt.text(x, y + 0.05, "%.2f" % y, ha = "center", va = "bottom", fontsize=14)
    plt.ylim(-1.25, +1.25)
    
for x, y in zip(X, Y2):
    plt.text(x, -y - 0.1, "%.2f" % y, ha = "center", va = "bottom", fontsize=14)
    # Correcto -> plt.text(x, -y - 0.05, "%.2f" % y, ha = "center", va = "top", fontsize=14)
    plt.ylim(-1.25, +1.25)
    
plt.show()
```

```python
plt.figure(4, figsize=[15, 10])

Z = np.full(20, 20)
Z[len(Z)-1] = 40
explodeArray = np.full(20, 0.04)
explodeArray[len(explodeArray)-1] = 0.1
colorsArray = plt.cm.gray(np.linspace(0,1,20))

plt.pie(Z, explode = explodeArray, colors = colorsArray,
        wedgeprops = { "edgecolor" : "black", 'linewidth': 1})

plt.show()
```
